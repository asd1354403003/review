```python
example = pd.read_csv('100.csv')
```

```python
# content-> x  sentiment-> y
xiao = example[['content', 'sentiment']]
```

```python
stopwords_path = 'stopwords.txt' # 停用词词表
```

```python
# 分词的时候 dask delayed =》这样可以快一点 因为 parallelly 计算
@delayed

# 分词 官网照抄的  
def delayed_jiebaclearText(text):
    mywordlist = []
    seg_list = jieba.cut(str(text), cut_all=False)
    liststr="/ ".join(seg_list)
    f_stop = open(stopwords_path)

    with open('stopwords.txt', 'r', encoding='UTF-8') as f1:
        f_stop_text = f1.read()    # 读取数据

    
    f_stop_seg_list=f_stop_text.split('\n')
    for myword in liststr.split('/'):
        if not(myword.strip() in f_stop_seg_list) and len(myword.strip())>1:
            mywordlist.append(myword)
    return ''.join(mywordlist)
```

```python
# 取6k个
duohsao = 6000
```

```python
li = []
# 到第 6k 个
for i in xiao.iloc[:duohsao, :].content:
    # append 每一个 jieba
    li.append(delayed_jiebaclearText(i))
```

```python
# 所有的delay 成一个 delayed的object 好计算
total_pic = dask.delayed(li)
```

```python
# 因为 加了delayed的装饰器之后都是 delayed 的object
# 所以要算出真实的数要 compute
total = total_pic.compute(num_workers=2)
```

```python
# 前面的好是所有在一起的 东西
# 所以 -duoshao 开始 才是真正的第一个数据
train = total[-duohsao:]
```

```python
# 所有的词都 转成 有序的的tensor（包含出现次数）
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
```

```python
# fit trainning data 
X = vectorizer.fit_transform(train)
```

```python
# .long() 是都转成 int
train = torch.Tensor(X.toarray()).long()
```

```python
# 它说不能有 -1 这种数字。 所以都加一
y = torch.Tensor(xiao.sentiment.values).long() +1
```

```python
# split to train， val， test
x_train, x, y_train, y = train_test_split(train, y[:duohsao], test_size=0.2)
x_val, x_test, y_val, y_test = train_test_split(x, y, test_size=0.5)
```

```python
x_train.shape
```

```python
# data loader-以batch size 喂数据
train_dataset = Data.TensorDataset(x_train, y_train)
val_dataset = Data.TensorDataset(x_val, y_val)

train_dataloader = Data.DataLoader(
    dataset=train_dataset,
    batch_size=batch_size,
    shuffle=True,
)
valid_dataloader = Data.DataLoader(
    dataset=val_dataset,
    batch_size=batch_size,
    shuffle=True,
)
```

```python
def train(dataloader, model, criterion, optimizer, device):

    model.train()
    epoch_losses = []
    epoch_accs = []
    
    # batch_x，batch_y trainloader 的 x 和 y
    for step, (batch_x, batch_y) in enumerate(dataloader):
        # 转到 gpu
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)
        
        # 带入 model
        prediction = model(batch_x.long())
        
        # 用 criterion 算损失
        loss = criterion(prediction, batch_y)
        
        # 算 accuracy
        accuracy = get_accuracy(prediction, batch_y)
        
        # model optimizer的参数梯度初始化为0
        model.zero_grad()
        optimizer.zero_grad()
        
        #将当前GPU储存gradient的计算图反传
        loss.backward()
        
        #利用计算图中储存的gradient更新网络参数
        optimizer.step()
        
        # append loss 和 accuracy
        epoch_losses.append(loss.item())
        epoch_accs.append(accuracy.item())

    return epoch_losses, epoch_accs
```

```python
def evaluate(dataloader, model, criterion, device):
    
    model.eval()
    epoch_losses = []
    epoch_accs = []
    
    # 数据不需要计算梯度，也不会进行反向传播
    with torch.no_grad():
        
        #下面的跟上面的一样
        for step, (batch_x, batch_y) in enumerate(dataloader):
            batch_x, batch_y = batch_x.to(device), batch_y.to(device)    
            prediction = model(batch_x)
            loss = criterion(prediction, batch_y)
            accuracy = get_accuracy(prediction, batch_y)
            epoch_losses.append(loss.item())
            epoch_accs.append(accuracy.item())
    return epoch_losses, epoch_accs
```

```python
def get_accuracy(prediction, label):
    # get size 
    batch_size, _ = prediction.shape
    # get predicted
    predicted_classes = prediction.argmax(dim=-1)
    # get correct_predictions
    correct_predictions = predicted_classes.eq(label).sum()
    # get accuracy
    accuracy = correct_predictions / batch_size
    return accuracy
```

```python
# 其实有很多参数还没用。因为都直接在layer里面直接改的数据。。


class CNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, n_filters, filter_sizes, output_dim, dropout_rate, stride):
        
        # 固定的 样子 super(CNN, self).__init__()
        super().__init__()
        
        # 卷积层 in_channels = 1, out_channels = 1, filter=3
          # 因为我就一层 卷积，所以我选 out_channels = 1.
        self.con1 = nn.Conv1d(1, 1, 3)
       
        # max_pooling, 我选3的时候 都是 0.88； 选 2 好了很多. 我之前是avgpooling 比max 差些
        self.max_pool = nn.MaxPool1d(2, stride=stride) 

        # Linear 我直接写的 3381 which 是池化后的shape; 应该有其他的直接查的表示，但是为了简单我就直接带数字了。
        # output_dim =》 3 因为 有 -1 0 1 三个 result
        self.fc = nn.Linear(3381, output_dim)
        # drop out layer
        self.dropout = nn.Dropout(dropout_rate)  
        
    def forward(self, ids):
        
        # 转float 我没记错的话 应该是 unsqueeze 要float
        ids = ids.float()
        
        # 别人都是先一个tokenizer + embedding layer， 我没有。 因为我首先把数据转成tensor了， 然后内存也不够
            # 然后 别人转后的shape 应该是  （batch_size(48), embedding_size, 词的总个数）
        
        # 搞完之后的 shape 是 （batch_size(48), 1, 词的总个数）
        embedded = ids.unsqueeze(0).permute(1,0,2)
        
        # 卷积+池化+dropout+linear
        conved = self.con1(embedded)
        pooled = self.max_pool(conved)
        cat = self.dropout(pooled)
        cat = torch.squeeze(cat,1)
        prediction = self.fc(cat)

        return prediction

# 很多的都没用 
vocab_size = x_train.shape[1]
embedding_dim = 300
n_filters = 1
filter_sizes = [3,5,7]
output_dim = 3
dropout_rate = 0.25
stride = 5
model = CNN(vocab_size, embedding_dim, n_filters, filter_sizes, output_dim, dropout_rate, stride)

model
```

```python
# 设置 criterion, optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(params=model.parameters())
n_epochs = 15
best_valid_loss = float('inf')

# to gpu
model = model.to(device)
criterion = criterion.to(device)


train_losses = []
train_accs = []
valid_losses = []
valid_accs = []


for epoch in range(n_epochs):
    
    # 算train_loss, train_acc, valid_loss, valid_acc 
    train_loss, train_acc = train(train_dataloader, model, criterion, optimizer, device)
    valid_loss, valid_acc = evaluate(valid_dataloader, model, criterion, device)
    train_losses.extend(train_loss)
    train_accs.extend(train_acc)
    valid_losses.extend(valid_loss)
    valid_accs.extend(valid_acc)

    epoch_train_loss = np.mean(train_loss)
    epoch_train_acc = np.mean(train_acc)
    epoch_valid_loss = np.mean(valid_loss)
    epoch_valid_acc = np.mean(valid_acc)
    
    if epoch_valid_loss < best_valid_loss:
        best_valid_loss = epoch_valid_loss
    
    print(f'epoch: {epoch+1}')
    print(f'train_loss: {epoch_train_loss:.3f}, train_acc: {epoch_train_acc:.3f}')
    print(f'valid_loss: {epoch_valid_loss:.3f}, valid_acc: {epoch_valid_acc:.3f}')
    
#     if epoch_valid_acc>= 0.9:
#         break
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
ax.plot(train_losses, label='train loss')
ax.plot(valid_losses, label='valid loss')
plt.legend()
ax.set_xlabel('updates')
ax.set_ylabel('loss');
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
ax.plot(train_accs, label='train accuracy')
ax.plot(valid_accs, label='valid accuracy')
plt.legend()
ax.set_xlabel('updates')
ax.set_ylabel('accuracy');
```
